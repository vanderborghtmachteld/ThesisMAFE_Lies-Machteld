```python
import pandas as pd
import numpy as np
import tensorflow as tf
import seaborn as sns
from keras import layers, models
from keras.models import Sequential
import os
from matplotlib import pyplot as plt

# ====================
# 1) Read your dataset
# ====================
os.chdir("..")
MYwd= os.getcwd()
file_path = MYwd+r"\Data\cause_specific_Short_Reconstructed\USA_m_short_idr.csv"
data = pd.read_csv(file_path)

# dividy by 1,000,000 
exclude_cols = ['country', 'year', 'sex', 'list', 'agf', 'cause']
data.loc[:, ~data.columns.isin(exclude_cols)] = data.loc[:, ~data.columns.isin(exclude_cols)] / 1000000

# only keep the lines for both sexes: 
subset = data[data['sex'] == 3]
subset = subset.drop(columns=['sex', 'list', 'agf', 'total', 'm85p', 'm90p', 'm95p'])

df = subset.melt(id_vars=['country', 'year', 'cause'], var_name='age', value_name='m')
df['age'] = df['age'].str.extract('(\d+)').astype(int)                           # Convert 'age' from 'mX' format to just 'X'
df['log_m'] = np.log(df['m'].replace(0,1e-8))                                    # add log mortality, avoid taking log(0)

# Prepare data to feed into neural network: 
df['cause'] = df['cause'].astype('category').cat.codes      # convert to categorical variable

df['year'] = df['year'].astype('float')    # neural nets prefer floats over ints
df['age'] = df['age'].astype('float')
# df has columns: ["country", "year", "cause", "age", "m", "log_m"]

# =============================

train = df[(df['year'].between(1979, 2011))]
train = train.copy() 

test = df[(df['year'].between(2012, 2021))]
test = test.copy()

# =============================

df_pivot_train = train.pivot_table(index=['age', 'year'], columns='cause', values='log_m').reset_index()
df_pivot_test = test.pivot_table(index=['age', 'year'], columns='cause', values='log_m').reset_index()

ages = df["age"].unique()
```

```python
num_sequences = len(ages)                        # 22
T_train = len(train["year"].unique())            # 33
T_test = len(test["year"].unique())              # 10
num_causes = len(train["cause"].unique())        # 17
```

```python
def create_X_y(df_pivot, T):
    X_list = []
    y_list = []
    
    # Loop over each unique age to form a sequence.
    for age in df_pivot['age'].unique():
        # Select rows corresponding to the current age and sort by year.
        df_age = df_pivot[df_pivot['age'] == age].sort_values(by='year')
        
        # Check if this age group has the full set of timesteps.
        if len(df_age) == T:
            # Build the input sequence.
            # Here, we use only the "age" column as the feature.
            # (Optionally, you might include 'year' or other features, e.g., df_age[['age', 'year']])
            seq = df_age[['age']].values  # Shape will be (T, 1)
            X_list.append(seq)
            
            # Build the target vector.
            # For a many-to-one setup, we choose the final row’s cause columns.
            # We remove the 'age' and 'year' columns, assuming the rest are causes.
            target = df_age.drop(columns=['age', 'year']).values[-1]  # Shape: (num_causes,)
            y_list.append(target)
    
    # Convert lists to numpy arrays.
    X = np.array(X_list)  # Shape: (num_age_groups, T, 1)
    y = np.array(y_list)  # Shape: (num_age_groups, num_causes)
    return X,y
```

```python
X_train, y_train = create_X_y(df_pivot_train, T_train)
X_test, y_test = create_X_y(df_pivot_test, T_test)
```

```python
num_input_features = 1  # e.g. just "age" at each year

# --------------------------------------------------------------
# 2) BUILD THE MODEL
# --------------------------------------------------------------

model = models.Sequential()

# The Input layer: we have T timesteps and num_input_features features
model.add(layers.Input(shape=(T_train, num_input_features)))

# An LSTM layer with 64 units (you can adjust as needed)
model.add(layers.LSTM(64))

# Final Dense layer with num_causes outputs (linear activation)
model.add(layers.Dense(num_causes, activation='linear'))

learning_rate = 0.005
optimizer = tf.keras.optimizers.Adam(learning_rate)
model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

model.summary()


# --------------------------------------------------------------
# 3) TRAIN THE MODEL
# --------------------------------------------------------------
history = model.fit(X_train, y_train, 
          epochs=400,      # increase as needed
          batch_size=22,  # choose an appropriate batch size
          validation_split=0.2,
          verbose=0)
```

```python
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
def forecast_recursive_batch(model, input_batch, forecast_steps=10):
    """
    Recursively forecast future outputs in batch for models that take constant input features.
    
    Args:
      model: A trained Keras model.
      input_batch: NumPy array of shape (num_age_buckets, T, num_features), 
                   e.g. (22, 33, 1) where each timestep contains a constant age value.
      forecast_steps: Integer. The number of future timesteps to forecast.
    
    Returns:
      forecasts: NumPy array of shape (forecast_steps, num_age_buckets, num_causes)
    """
    forecasts = []
    # Copy the batch input so that updates do not modify the original data.
    batch_seq = input_batch.copy()  # shape: (22, 33, 1)
    
    for _ in range(forecast_steps):
        # Model expects input shape (batch_size, timesteps, num_features)
        pred = model.predict(batch_seq)  # pred will have shape: (22, num_causes)
        forecasts.append(pred)
        
        # Update each sequence in the batch:
        # Remove the first timestep, and then append a new one.
        # Since the only feature is age (which is constant), 
        # the new timestep is simply the same constant value as the last timestep.
        new_timestep = batch_seq[:, -1, :]  # shape: (22, 1)
        new_timestep = new_timestep.reshape(batch_seq.shape[0], 1, batch_seq.shape[2])
        batch_seq = np.concatenate([batch_seq[:, 1:, :], new_timestep], axis=1)
    
    forecasts = np.array(forecasts)  # shape: (forecast_steps, 22, num_causes)
    return forecasts
```

```python
# --------------------------------------------------------------
# 4) PREDICT
# --------------------------------------------------------------

# Define parameters
forecast_steps = 10  # For example, forecast for the next 10 years
years = np.arange(1979+forecast_steps, 1979+T_train+forecast_steps)

input_batch = np.zeros((len(ages), T_train, 1), dtype=np.float32)
for i, age in enumerate(ages):
    # For each bucket, every timestep has the constant value of the bucket's age.
    input_batch[i, :, 0] = age

predicted_future = forecast_recursive_batch(model, input_batch, forecast_steps=forecast_steps)
predicted_future = np.transpose(predicted_future, (1, 0, 2))

test['prediction'] = predicted_future.flatten()
print("Shape of forecasted predictions:", predicted_future.shape)
```

```python
# goodness of fit on plot
g = sns.FacetGrid(test, col='cause', palette='mako',
                  margin_titles=True, height=4, aspect=0.7, col_wrap=6)

g.map(sns.scatterplot, 'age', 'log_m', s=20)

# Adjust the title, legend, and axis formatting
g.set_axis_labels("Age", "Log Mortality")


for ax, (_, subset) in zip(g.axes.flat, test.groupby(['cause'])):
    scatter = ax.scatter(subset['age'], subset['log_m'], label='actual')
    ax.scatter(subset['age'], subset['prediction'], label='prediction')
    ax.legend()
```
